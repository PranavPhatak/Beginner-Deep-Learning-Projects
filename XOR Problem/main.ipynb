```python
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
```

```python
y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x * (1-x)
```

```python
h1 = 2
h2 = 2
w1 = np.random.randn(2, h1)
b1 = np.zeros((1, h1))

w2 = np.random.randn(h1, h2)
b2 = np.zeros((1, h2))

w3 = np.random.randn(h2, 1)
b3 = np.zeros((1, 1))

lr = 0.1
```

```python
for epoch in range(10000):
    hidden1 = sigmoid(np.dot(X, w1) + b1)
    hidden2 = sigmoid(np.dot(hidden1, w2) + b2)
    output = sigmoid(np.dot(hidden2, w3) + b3)

    loss = np.mean((y - output) ** 2)

    output_error = y - output
    output_delta = output_error * sigmoid_derivative(output)

    hidden2_error = output_delta.dot(w3.T)
    hidden2_delta = hidden2_error * sigmoid_derivative(hidden2)

    hidden1_error = hidden2_delta.dot(w2.T)
    hidden1_delta = hidden1_error * sigmoid_derivative(hidden1)

    w3 += hidden2.T.dot(output_delta) * lr
    b3 += np.sum(output_delta, axis=0, keepdims=True) * lr

    w2 += hidden1.T.dot(hidden2_delta) * lr
    b2 += np.sum(hidden2_delta, axis=0, keepdims=True) * lr

    w1 += X.T.dot(hidden1_delta) * lr
    b1 += np.sum(hidden1_delta, axis=0, keepdims=True) * lr

print("Predicted Output: \\n", output)
print("Loss after training:\\n", loss)
```
